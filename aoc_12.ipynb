```julia
I = only.(mapreduce(permutedims, vcat, split.(readlines(open("input")), "")))
```

```julia
M = zeros(Int, size(I))
```

```julia
inbounds(i) = 1 <= i[1] <= size(I,1) && 1 <= i[2] <= size(I,2)
up, right, down, left = CartesianIndex(-1, 0), CartesianIndex(0,1), CartesianIndex(1,0), CartesianIndex(0,-1)
unvisited_same_neighbor(i,v) = [i+d for d in [up, right, down, left] if inbounds(i+d) && I[i+d]==v && !((i+d) in visited)]
```

```julia
function floodfill(i, c)
    M[i] = c
    push!(visited, i)
    [floodfill(n, c) for n in unvisited_same_neighbor(i, I[i])]
end
```

```julia
visited = []
c = 1
for i in CartesianIndices(I)
    (i in visited) && continue
    floodfill(i, c)
    c += 1
end
```

```julia
fence(i) = sum([!inbounds(i+d) || I[i+d] != I[i] for d in [up, right, down, left]])
```

```julia
fences(v) = sum(fence(i) for i in findall(==(v), M))
```

```julia
price(v) = fences(v) * count(==(v), M)
```

```julia
sum(price.(1:maximum(M)))
```

```julia
function h_edgefill!(drawn, i, v, D)
    drawn[i][D] = true
    d = left
    while (!inbounds(i+d+D) || M[i+d+D] != v) && inbounds(i+d) && M[i+d] == v 
        drawn[i+d][D] = true
        d += left
    end
    d = right
    while (!inbounds(i+d+D) || M[i+d+D] != v) && inbounds(i+d) && M[i+d] == v 
        drawn[i+d][D] = true
        d += right
    end
end

function v_edgefill!(drawn, i, v, D)
    drawn[i][D] = true
    d = up
    while (!inbounds(i+d+D) || M[i+d+D] != v) && inbounds(i+d) && M[i+d] == v 
        drawn[i+d][D] = true
        d += up
    end
    d = down
    while (!inbounds(i+d+D) || M[i+d+D] != v)  && inbounds(i+d) && M[i+d] == v 
        drawn[i+d][D] = true
        d += down
    end
end
```

```julia
function n_sides(v)
    all_idxs = findall(==(v), M)
    drawn = Dict(i=>Dict(up=>false, down=>false, left=> false, right=>false) for i in all_idxs)
    c = 0
    for i in all_idxs
        if (!inbounds(i+up) || M[i+up] != v) && !drawn[i][up]
            h_edgefill!(drawn, i, v, up)
            c += 1
        end
        if (!inbounds(i+down) || M[i+down] != v) && !drawn[i][down]
            println("d", i)
            h_edgefill!(drawn, i, v, down)
            c += 1
        end
        if (!inbounds(i+left) || M[i+left] != v) && !drawn[i][left]
            v_edgefill!(drawn, i, v, left)
            c += 1
        end
        if (!inbounds(i+right) || M[i+right] != v) && !drawn[i][right]
            v_edgefill!(drawn, i, v, right)
            c += 1
        end
    end
    return c
end
```
